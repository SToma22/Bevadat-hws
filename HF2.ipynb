**b)** (3 pont)<br>
**Osztályozás I: OOP**<br>
Első feladatunk a kNN algoritmus objektumorientált leprogramozása. Célunk, hogy egy olyan *MykNN* classt hozzunk létre, mely az sklearn-es osztályozóhoz hasonlóan használható .fit() és .predict() metódusokkal. Nem célunk, hogy az sklearnhez hasonló teljesítményű, gyorsaságú algoritmust implementáljunk, elegendő az egyszerű lazy kNN implementálása.
- Írjuk meg az osztály konstruktorát! A konstruktorának két bemenete legyen: *k*, azaz hány legközelebbi szomszédot vizsgáljunk, és *metric*, azaz milyen távolságfogalmat használjon az algoritmus. Ezeket tárolja el az objektum *k* és *metric* adattagokban.<br>
- Írjuk meg az osztály .fit() metódusát. A metódusnak két bemenete legyen X_train és y_train. Tételezzük fel, hogy a bemenet pandas DataFrame, illetve pandas Series. Mivel az algoritmus lazy verzióját szeretnénk implementálni, ezért elegendő, ha a metódus mindössze annyit csinál, hogy eltárolja X_train és y_train adattagokban a bemeneteket numpy arrayként. (A pandas DataFrameből könnyen numpy arrayt készíthetünk a pandas DataFramek *values* adaattagának segítségével. A biztonság kedvéért másoljuk le a DataFrameket a .copy() metódussal.)
- Írjuk meg az osztály .predict() metódusát. A metódusnak egy bemenete legyen X_test, kimenete pedig egy y_pred numpy array legyen, ami megadja, hogy az adott sornak a .metric metrika szerinti .k legközelebb szomszédját nézve milyen címkét kellene kapnia. A program ezen részének működését kommentek és/vagy szöveges válasz formájában **magyarázd is!**<br>

Miután implementáltuk az algoritmusunkat, próbáljuk is ki, hogy hogyan teljesít. Az sklearn.model_selection.train_test_split() függvényével osszuk fel véletlenszerűen 70% tanító és 30% teszthalmaz részre a red_wines táblázatunkat. Ne feledkezzünk meg egy random_state beállításáról, hogy össze tudjuk vetni majd más algoritmusok teljesítményével is a későbbiekben. <br>
Az sklearn.metrics accuracy_score(), precision_score(), recall_score() függvényei segítségével vizsgáld meg a mérőszámokat egy általad választott k és metrika esetén.

**Megjegyzés:** Tételezzük fel, hogy a metrika a scipy.spatial.distance.minkowski() függvényéhez hasonlóan működik. <br>
A k legközelebbi szomszéd kiszámítására nyugodtan alkalmazhatjuk a brute force algoritmust: Nézzük meg az összes pont távolságát az adott ponttól, majd vegyük a k legközelebbit! <br>

```python
class MykNN:
    def __init__(self, k, metric):
        """
         k: vizsgálandó legközelebbi szomszédok száma, egy természetes szám
         metric: amilyen metrikában nézzük a távolságot, természetes szám vagy szöveg
        """
        self.k = k
        self.metric = metric
    
    def fit(self, X_train, y_train):
        """
         Eltárolja self.X_train és self.y_train változókban
         input:
            X_train: pd.DataFrame, a tanító halmazunk
            y_train: pd.Series, a tanító halmazon felvett borminőségi értékek
        """
        self.X_train = X_train.copy().to_numpy()
        self.y_train = y_train.copy().to_numpy()
        #return None
    
    def predict(self, X_test):
        """
         input:
            X_test: DataFrame, a teszt halmazunk, nem tudjuk a bor minőségét de prediktálni szeretnénk
         input:
            y_pred: np.array(), az adott sor metrika szerinti k legközelebbi szomszédját nézve prediktáljon
        """
        x_test = X_test.copy().to_numpy()
        y_pred = np.array([0 for _ in range(len(X_test))])
        
        for i in range(len(x_test)):                                    ### végigmegyünk a teszthalmazon index-el
            d = {}                                                      ### távolság szótárja {'távolság':'prediktált értékek listája'} formában
            for j in range(len(self.X_train)):                          ### megnézzük minden pont távolságát a teszthalmaz i-edik sorával (adatával), itt j egy lista (az adott)
                sum = 0
                for k in range(len(x_test[0])):                         ### itt végigmegyünk az adott sor minden tulajdonságán, és megnézzük a sor távolságát a tesztelemre
                    sum += (x_test[i][k] - self.X_train[j][k]) ** self.metric
                sum ** 1/(self.metric)                                  ### megkaptuk a testhalmaz i-edik sorának távolságát train halmaz j-edik sorával
                if sum in d:
                    d[sum].append(self.y_train[j])
                else:
                    d[sum] = [self.y_train[j]]

            Keys = list(d.keys())
            Keys.sort()
            sorted_dict = {sd: d[sd] for sd in Keys}                    ### sorba rendeztük a szótárt, így a legközelebbi értékek vannak a legelöl
            lista = list(sorted_dict.values())
            jobor = 0
            rosszbor = 0
            
            for x in range(self.k):
                for y in lista[x]:
                    if y == 1:
                        jobor += 1
                    else:
                        rosszbor += 1
            if jobor >= rosszbor:
                y_pred[i] = 1
            else:
                y_pred[i] = 0
        return y_pred
```

```python
import sklearn.model_selection, sklearn.metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree


X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(red_wines.drop('quality', axis=1), red_wines.quality, test_size=0.3, random_state=0)
```

```python
körte = MykNN(5, 2)
körte.fit(X_train, y_train)
y_pred = körte.predict(X_test)
```

```python
print(y_pred)

print(f'Accuracy: {round(sklearn.metrics.accuracy_score(y_test, y_pred),2)}')
print(f'Precision: {round(sklearn.metrics.precision_score(y_test, y_pred),2)}')
print(f'Recall: {round(sklearn.metrics.recall_score(y_test, y_pred),2)}')
```

> Tipp: Használjuk a model_selection.cross_val_score() függvényt scoring = 'f1' beállítással! Boxplot készítéshez pedig a seaborn boxplot() függvényét!

```python
from sklearn import model_selection
```

```python
dict = {}
for i in range(1,10):
    szomszed = KNeighborsClassifier(n_neighbors=i, metric="euclidean")
    szomszed.fit(X_train, y_train)

    scores_knn = model_selection.cross_val_score(szomszed, X_train, y_train, scoring='f1', cv = 10)
    dict[i] = scores_knn

sns.boxplot(dict)
#for j in dict:
#    print(f"{j} darab szomszédot vizsgálva az adathalmazunkon a modell pontossága: {dict[j]}")
```

```python
fa_melyseg = {}
for i in range(10, 23):
    fa = tree.DecisionTreeClassifier(max_depth=i)
    fa = fa.fit(X_train, y_train)

    scores_fa = model_selection.cross_val_score(fa, X_train, y_train, scoring='f1', cv = 10)
    fa_melyseg[i] = scores_fa

sns.boxplot(fa_melyseg)

#for j in fa_melyseg:
#    print(f"{j} szint mély döntési fát készítve a modell pontossága: {fa_melyseg[j]}")
```
